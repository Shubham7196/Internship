## Q - 1. Scrape the details of most viewed videos on YouTube from Wikipedia

```python
import pandas as pd
import numpy as np
import selenium
import time
from bs4 import BeautifulSoup 
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import requests
import re
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver = webdriver.Chrome(r'C:\Users\Wystre\Desktop\chromedriver.exe')
driver.get(url)

```

```python
rank = []
name = []
artist = []
date = []
views = []

for k in range(1,31):
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[1]".format(k)):
            rank.append(i.text)
    except NoSuchAttributeException as e:
        rank.append('--')
        
    
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[2]".format(k)):
            name.append(i.text)
    except NoSuchAttributeException as e:
        name.append('--')
        
        
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[3]".format(k)):
            artist.append(i.text)
    except NoSuchAttributeException as e:
        artist.append('--')
        
        
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[4]".format(k)):
            views.append(i.text)
    except NoSuchAttributeException as e:
        views.append('--')
        
        
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[5]".format(k)):
            date.append(i.text)
    except NoSuchAttributeException as e:
        date.append('--')
```

```python
print(len(rank))
print(len(name))
print(len(artist))
print(len(date))
print(len(views))
```

```python
df=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Upload date':date,'Views':views})
df
```

## 5. Scrape the details of trending repositories on Github.com.

```python
url = "https://github.com/"
driver = webdriver.Chrome(r'C:\Users\Wystre\Desktop\chromedriver.exe')
driver.get(url)
```

```python
exp_btn = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
try:
    exp_btn.click()
except:
    pass
```

```python
trnd_url = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
urls = trnd_url.get_attribute("href")
driver.get(urls)
```

```python
# Creating Empty list

repo_urls = []
rep_title = []
description = []
contributors = []
language = []
lang = []
# Fetching URLs for each repository

repo = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']//a")
for i in repo:
    repo_urls.append(i.get_attribute("href"))
    
#Scrapping data of Repository Title

title = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
for i in title:
    rep_title.append(i.text)
    
#Scrapping data from every Repository page:
for i in repo_urls:
    driver.get(i)
    time.sleep(3)

    #Scrapping data of description:
    try:
        desc = driver.find_element_by_xpath("//p[@class = 'f4 my-3']")
        description.append(desc.text)
    except:
        description.append('-')
        
    #SCrapping data of Contributors:
    try:
        cont = driver.find_element_by_xpath("//a[contains(@href,'contributors')]/span")
        contributors.append(cont.text)
    except:
        contributors.append('-')
    
    #Scrapping language used:
    try:
        for i in driver.find_element_by_xpath("//a[contains(@href,'search?')]/span[1]"):
            lang.append(i.text)
        language.append(lang)
    except:
        language.append('-')
```

```python
gith = {"Repository Title":rep_title,"Repository Description":description,"Contributors Counts":contributors,"Language Used":language}
GitHub = pd.DataFrame(gith)
GitHub.reset_index(drop = True, inplace = True)
GitHub.index+=1
```
